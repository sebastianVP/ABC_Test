### ¿Qué es un BoxPlot?

Un BoxPlot (o diagrama de caja, en traducción libre) muestra la distribución cuantitativa de los datos de una manera que facilita la comparación entre las variables, o a través de los niveles categóricos de las variables.

Esta caja ("box") muestra los cuartiles del conjunto de datos mientras que los "whiskers" muestran el resto de la distribución, excepto los puntos que se denominan valores atípicos.

cuartil.png

La línea en el centro de la figura en verde, en nuestro caso, representa la mediana. La línea azul que está en el borde izquierdo de la figura representa el 25 % de mi distribución, y la línea que está en el borde derecho representa el 75 % de la distribución de datos.

En las líneas moradas que están en los extremos de ambos lados, dejan la figura central y muestran una delimitación, tenemos el inicio y el final del área de los cuartiles. Ahora, observe que hay algunos puntos después de este límite en el lado izquierdo. Estos puntos se denominan valores atípicos, es decir, son puntos “aislados” en nuestra distribución

## Codificación categórica

Este es el proceso de transformar los datos de texto (datos bajo variable categórica) en números para que el modelo pueda comprenderlos y extraer información valiosa de ellos. Estas son algunas técnicas poderosas de codificación categórica:

* Codificación ordinal: implica la asignación de cada etiqueta única a un valor integral. Este tipo de codificación es aconsejable sólo si existe un orden/relación conocido entre las categorías. P.ej. Tomemos un ejemplo de una variable de calidad del producto que tiene valores:

  "Excelente", "Muy buena", "Buena", "Promedio" y "Deficiente".

  Dado que estos datos tienen un orden, es decir, 'Excelente '>'Muy bueno'>'Bueno'>'Promedio'>'Deficiente', podemos usar la codificación ordinal para asignar:
 'Deficiente': 0, 'Promedio': 1, 'Bueno ': 2, 'Muy bueno': 3, 'Excelente': 4.

* One Hot Encoding: esta es la transformación de datos en valores binarios en función de la aparición de categorías. Esto se utiliza sólo cuando no hay orden/relación entre las categorías. Por ejemplo, en nuestros datos, tenemos una variable llamada Género que consta de valores: Masculino y Femenino. Como no hay orden, utilizamos One Hot Encoding.

## Lidiar con los valores perdidos (DEALING WITH MISSING VALUES)
Algunos de los métodos que se utilizan para manejar valores nulos en los datos son:

* Imputación de muestra aleatoria: implica tomar observaciones aleatorias del conjunto de datos y usarlas para reemplazar valores nulos. Es apropiado cuando  faltan datos completamente al azar (MCAR).
* Imputar media/mediana: este es el método de completar los datos faltantes con la media/mediana de su atributo respectivo. Es apropiado cuando faltan datos al azar (MAR).
* Predicción de los valores nulos: cuando existe una alta colinealidad entre el atributo con valores nulos y el que no tiene valores nulos, es adecuado utilizar la regresión lineal para predecir y completar los valores faltantes.


```python
df.isnull().sum()
```

* Los valores nulos en todas las columnas faltan al azar. Probablemente los empleados no quisieron compartir o dudaron en compartir información como la puntuación de estrés, la tasa de quemado y las horas de trabajo.
* Faltan valores en la puntuación de tasa de quemado y fatiga mental. Visualicemos su distribución.


```python
sns.heatmap(df.isnull())
```

Observamos que los valores nulos en las dos características no ocurren simultáneamente (solo 172 valores ocurren simultáneamente) para una muestra determinada. Por lo tanto, imputamos el valor faltante del objetivo mediante la puntuación de fatiga mental y viceversa mediante la regresión lineal debido a su fuerte colinealidad. Usaremos la imputación de mediana/media para el resto de los valores faltantes (donde los valores nulos ocurren simultáneamente).

```python
x1 = pd.DataFrame(df[df['Burn Rate'].isnull()]['Mental Fatigue Score'])
```

```python
x1['Mental Fatigue Score'].fillna(x1['Mental Fatigue Score'].median(),inplace=True)
#Filling the simultaneously occuring null values for Mental Fatigue Score with median imputation.
```

```python
#Imputation using Linear Regression
df_new = df[['Mental Fatigue Score','Burn Rate']] #the training data with no null values
df_new.dropna(inplace=True)
```

Usar la puntuación de fatiga mental como característica independiente y la tasa de quemado como característica dependiente en el modelo de regresión lineal.

```python
X = df_new[['Mental Fatigue Score']] # X.shape should be (N, M) where M >= 1
y = df_new['Burn Rate']
model = LinearRegression()
model.fit(X,y)
pred = model.predict(x1)
#the values to be filled in nan values of burn rate
ind = df[df['Burn Rate'].isnull()].index
for j,i in enumerate(ind):
 df['Burn Rate'].iloc[i] = pred[j]
#Filling the missing values with   the predicted values
```

Es hora de eliminar los valores atípicos que detectamos en algunas funciones durante EDA mediante la visualización de diagramas de caja.

Los valores atípicos son puntos de los datos que se alejan mucho de la otra observación. La presencia de valores atípicos puede estropear y engañar el proceso de entrenamiento del modelo. De ahí la importancia de manejarlos.

Aquí usaremos log transformations en los datos para deshacernos de los valores atípicos. Usaremos este método en las entidades con un módulo de asimetría >0,5, ya que los datos con una asimetría inferior a 0,5 siguen una distribución bastante simétrica.


```python
import numpy as np
```

# NOTA SOBRE PICKLE

El módulo Pickle de Python es un popular formato utilizado para serializar y deserializar tipos de datos. Este formato es nativo de Python, lo que significa que los objetos Pickle no pueden cargarse utilizando ningún otro lenguaje de programación.

Pickle tiene sus propias ventajas e inconvenientes en comparación con otros formatos de serialización.

**Ventajas de utilizar Pickle para serializar objetos**


A diferencia de formatos de serialización como JSON, que no pueden gestionar tuplas y objetos datetime, Pickle puede serializar casi todos los tipos de datos integrados de Python de uso común. También conserva el estado exacto del objeto, cosa que JSON no puede hacer.
Pickle también es una buena opción para almacenar estructuras recursivas, ya que solo escribe un objeto una vez.
Pickle ofrece flexibilidad a la hora de deserializar objetos. Puedes guardar fácilmente distintas variables en un archivo Pickle y volver a cargarlas en otra sesión de Python, recuperando tus datos exactamente como estaban sin tener que editar tu código.


**Desventajas de utilizar Pickle**


Pickle no es seguro porque puede ejecutar callables Python maliciosos para construir objetos. Al deserializar un objeto, Pickle no puede distinguir entre un callable malicioso y uno no malicioso. Debido a esto, los usuarios pueden acabar ejecutando código arbitrario durante la deserialización.
Como ya se ha mencionado, Pickle es un módulo específico de Python, y puede que te cueste deserializar objetos pickled cuando utilices un lenguaje diferente.
Según varias pruebas comparativas, Pickle parece ser más lento y produce valores serializados mayores que formatos como JSON y ApacheThrift.
